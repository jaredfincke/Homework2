### I then pull the all open coffee shops in Catonsville, MD

```python

is_closed = "TRUE"
search_limit = 50
location = "Catonsville MD"
term = "Coffee"
url = 'https://api.yelp.com/v3/businesses/search'

headers = {
        'Authorization': 'Bearer {}'.format(api_key),
    }

url_params = {
                'Businesses.is_closed': is_closed,
                'location': location.replace(' ', '+'),
                'limit': search_limit,
                'term' : term
            }
response = requests.get(url, headers = headers, params = url_params)
print(response)
print(type(response.text))
print(response.text[:1000])
```

### Using the Pandas, I put the json into a dataframe. 

### I also add an column for Area.

```python
df = pd.DataFrame.from_dict(response.json()['businesses'])
# Declare a list that is to be converted into a column  
df['Area'] = 'Catonsville'
print(len(df)) 
print(df.columns) 
df.to_csv(r'../Data/raw_yelp_Cville.csv', index = False)
```

### I then pull the all open coffee shops in Dundalk, MD
